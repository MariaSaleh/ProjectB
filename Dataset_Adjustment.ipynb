```python
import pandas as pd
import re
import glob
import os
from os import path

# Help Functions

def read_file(file_path):
    if file_path.endswith(".xlsx"):
        xlsx = pd.ExcelFile(file_path)
        return xcel(xlsx, sheet_name='Sheet1')
    elif file_path.endswith(".csv"):
        return pd.read_csv(open(file_path, 'r'))
    
# Aux Functions

def file_add_date(file_path):
    p = re.compile('(.*(\\\\|\/))?(\d\d-\d\d-\d\d\d\d)\.(xlsx|csv)')
    m = p.match(file_path)
    if m:
        date = m.group(3)
    else:
        return None
    df = read_file(file_path)
    df.insert(0, "date", date, True)
    return df

# Main Functions

def adjust_data_base(data_dir, out_dir):
    add_header = True
    if not path.exists(out_dir+'/latest'):
        os.mkdir(out_dir+'/latest')
    if path.exists(out_dir+'/latest/all_daily.csv'):
        os.remove(out_dir+'/latest/all_daily.csv')
    f = open(out_dir+'/latest/all_daily.csv', 'a+')
    for file_path in glob.iglob(data_dir+'/*'):
        print(file_path)
        df = file_add_date(file_path)
        if df is None:
            continue
        if add_header:
            df.to_csv(f,index=False,line_terminator='\n')
            add_header = False
        else:
            df.to_csv(f,index=False,header=False,line_terminator='\n')    
    f.close()
    return 1


adjust_data_base('COVID-19\csse_covid_19_data\csse_covid_19_daily_reports','.')
```
